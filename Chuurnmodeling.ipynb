```python
# churn_mol['Gender_Male'] = churn_mol['Gender_Male'].astype(int)

# print(churn_mol)
# churn_mol['Gender_Male'] = churn_mol['Gender_Male'].replace({True: 1, False: 0})

# print(churn_mol)

boolean_cols = churn_mol.select_dtypes(include=['bool']).columns
churn_mol[boolean_cols] = churn_mol[boolean_cols].astype(int)
```

```python
from sklearn.model_selection import train_test_split
```

```python
  # Replace with your actual target column name
# IsActiveMember


X= churn_mod.drop(columns='IsActiveMember',axis=1)
y = churn_mod['IsActiveMember']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,stratify=y)
```

```python
churn_mod.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
prediction = model.predict(X_test)
```

```python
prediction
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

```python
accuracy = accuracy_score(y_test, prediction)
print(f'Accuracy: {accuracy:.2f}')




# Precision
precision = precision_score(y_test,prediction)
print(f'Precision: {precision:.2f}')

# Recall
recall = recall_score(y_test, prediction)
print(f'Recall: {recall:.2f}')

# # F1 Score
f1 = f1_score(y_test, prediction)
print(f'F1 Score: {f1:.2f}')

# # Confusion Matrix
conf_matrix = confusion_matrix(y_test, prediction)
print('Confusion Matrix:')
print(conf_matrix)

# # Classification Report
class_report = classification_report(y_test,prediction)
print('Classification Report:')
print(class_report)
```
